## Getting rid of outliers

```python
#plot using seaborn
plt.figure(figsize=(12, 6))
sns.boxplot(y=df['Total.Injuries'])  # y only since it's 1 variable
#plot
plt.ylabel('Total Injuries')
plt.xlabel('Total.Injuries')
plt.show()
```

As you can see from the above boxplot our column `Total.Injuries` which is a sum of all the injuries incured has a lot of outliers and inorder of us to get a normal distribution , we will need to get rid of the outliers.
Mean which is our main measure of central tendency will be highly affected by extreme values hence our predictions will be biased and distort our readings.

```python
#get the interquartile range
Q1 = df['Total.Injuries'].quantile(0.25)
Q3 = df['Total.Injuries'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR # using this standard formulae
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame
df = df[(df['Total.Injuries'] >= lower_bound) & (df['Total.Injuries'] <= upper_bound)]
```

```python
#check for the new column
df.describe()
```

Repeat the same code to remove outliers for the `Total.uninjured` column and `Number.of.Engines` column

```python
#get the interquartile range 
Q1 = df['Total.Uninjured'].quantile(0.25)
Q3 = df['Total.Uninjured'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR # using this standard formulae
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame
df = df[(df['Total.Uninjured'] >= lower_bound) & (df['Total.Uninjured'] <= upper_bound)]
```

```python
Q1 = df['Number.of.Engines'].quantile(0.25)
Q3 = df['Number.of.Engines'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR # using this standard formulae
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame
df = df[(df['Number.of.Engines'] >= lower_bound) & (df['Number.of.Engines'] <= upper_bound)]
```

## Comparison between 'Destroyed' vs 'Minor'

```python
destroyed_aircraft = df[df['Aircraft.damage'].isin(['Destroyed', 'Minor'])]
damage_counts = destroyed_aircraft.groupby(['Make', 'Aircraft.damage']).size().unstack(fill_value=0)

# Keep only the top 10 makes with most total incidents (Destroyed + Minor)
top_makes = damage_counts.sum(axis=1).sort_values(ascending=False).head(10).index
damage_counts = damage_counts.loc[top_makes]

# Step 4: Plot grouped bar chart
damage_counts.plot(kind='bar', figsize=(12, 6))
plt.title("Comparison of 'Destroyed' vs 'Minor' Damage by Aircraft Make")
plt.xlabel("Make")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45, ha='right')
plt.legend(title='Aircraft Damage')#show key
plt.show()
```
